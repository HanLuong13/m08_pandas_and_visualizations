```python
import pandas as pd
payments = pd.read_csv(R'restaurant\chefmozaccepts.csv')
places = pd.read_csv(R'restaurant\geoplaces2.csv')
```

```python
# 1. What are all of the supported payment methods for each place?
name_and_places = places.merge(payments, how = 'inner', on = 'placeID')
name_and_places[['name', 'Rpayment']]
```

```python
#2. What are the parking statuses for each place?
parkings = pd.read_csv(R'restaurant\chefmozparking.csv')
name_and_parking = places.merge(parkings, how = 'inner', on = 'placeID')
name_and_parking[['name', 'parking_lot']]
```

```python
# 3. What are the cuisines for each place?
cuisines = pd.read_csv(R'restaurant\chefmozcuisine.csv')
name_and_cuisines = places.merge(cuisines, how = 'inner', on = 'placeID')
name_and_cuisines[['name', 'Rcuisine']]
```

```python
# 4. What is the average rating, average food rating, average service rating, minimum food rating, minimum service rating, and maximum service rating for the restaurants by cuisine?

ratings = pd.read_csv(R'restaurant\rating_final.csv')
ratings_cuisines = ratings.merge(cuisines, how='inner', on='placeID')
ratings_cuisines = ratings_cuisines.dropna(subset=['rating', 'food_rating', 'service_rating', 'Rcuisine'])

cuisine_list = ratings_cuisines.Rcuisine.unique()

for cuisine in cuisine_list:
    print(cuisine, ratings_cuisines[ratings_cuisines.Rcuisine == cuisine].describe())

```

```python
# 5. What are the payment methods preferred by social drinkers?

user_profiles = pd.read_csv(R'restaurant\userprofile.csv')
user_payments = pd.read_csv(R'restaurant\userpayment.csv')
user_profiles_and_payments = user_profiles.merge(user_payments, how = 'inner', on = 'userID')
user_profiles_and_payments.Upayment[user_profiles_and_payments.drink_level == 'social drinker'].value_counts()

```

```python
# 6. What are the cuisines preferred by eco-friendly users?

user_cuisines = pd.read_csv(R'restaurant\usercuisine.csv')
user_profiles_and_cuisines = user_profiles.merge(user_cuisines, how = 'inner', on = 'userID')
user_profiles_and_cuisines.Rcuisine[user_profiles_and_cuisines.interest == 'eco-friendly'].value_counts()
```

```python
# 7. What payment methods are preferred by students?

user_profiles_and_payments.Upayment[user_profiles_and_payments.activity == 'student'].value_counts()
```

```python
# 8. What cuisines are preferred by Catholic students?
user_profiles_and_cuisines.loc[(user_profiles_and_cuisines['activity'] == 'student') & (user_profiles_and_cuisines['religion'] == 'Catholic'),['Rcuisine']]
```

```python
# 9. What are some of the demographics of the users rating Mexican service ratings a 2?
ratings_user_profiles = ratings.merge(user_profiles, how = 'inner', on = 'userID')
ratings_cuisines_user_profiles = ratings_user_profiles.merge(cuisines, how = 'inner', on = 'placeID')
ratings_cuisines_user_profiles.loc[(ratings_cuisines_user_profiles['Rcuisine'] == 'Mexican'), ['religion', 'marital_status']]
```
