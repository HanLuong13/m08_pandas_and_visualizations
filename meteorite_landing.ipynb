```python
import pandas as pd
meteor = pd.read_csv('Meteorite_Landings.csv')
meteor = meteor.dropna(subset=['year'])
# meteor.info()
```

```python
# Remove the nametype field
meteor.drop('nametype',inplace=True,axis=1)
meteor.head()
```

```python
# Clean the mass (g) field so that there's a default value of 0 where there is no mass listed.

meteor['mass (g)'].fillna(0,inplace=True)
len(meteor[meteor['mass (g)'].isna()])
```

```python
# Store each decade's data in its own data sheet in Excel using pandas. So 1990-1999, 2000-2009, 2010-2019, etc.

meteor['year'].astype(int)
years = meteor['year'].unique()
with pd.ExcelWriter('meteor_by_year.xlsx') as xlsxWriter:
    for year in years:
        if 1990 <= year <= 1999:
            meteor[meteor.year == year].to_excel(xlsxWriter,sheet_name='1990-1999')
        elif 2000 <= year <= 2009:
            meteor[meteor.year == year].to_excel(xlsxWriter,sheet_name='2000-2009')
        elif 2010 <= year <= 2019:
            meteor[meteor.year == year].to_excel(xlsxWriter,sheet_name='2010-2019')

xlsxWriter.save()
```

```python
# Format the date in MM/dd/yyyy with the time in hours:minutes:seconds AM/PM.

meteor = pd.read_csv('Meteorite_Landings.csv')
meteor = meteor.dropna(subset=['year'])
meteor['mass (g)'].fillna(0,inplace=True)
meteor['year'] = meteor['year'].astype(int).astype(str)

print(meteor['year'])

def func(string):
    string = '{}-01-01'.format(string)
    return string

meteor['year'] = meteor['year'].apply(func)

meteor['year']


print(meteor['year'],meteor['year'].astype('datetime64[ns]'))
# pd.to_datetime(meteor.year, format = '%Y')
```
