```python
books = pd.read_csv('books.csv')
books = books.dropna(subset = ['isbn', 'isbn13', 'language_code','original_publication_year'])
books['isbn13'] = books['isbn13'].astype('int64').astype('str') 

def func(string):
    string = '{}-{}-{}-{}-{}'.format(string[0:3],string[3:4],string[4:9],string[9:12],string[12:])
    return string

books['isbn13'] = books['isbn13'].apply(func)

books['isbn13']
```

```python
books.assign(difference_title = books['title'] == books['original_title'])
```

```python
books.original_publication_year.astype(int)

books_year = books.original_publication_year.unique()
with pd.ExcelWriter('books_by_year.xlsx') as xlsxWriter:
    for year in books_year:
        if 1800 <= year <= 1899: 
            books[books.original_publication_year == year].to_excel(xlsxWriter,sheet_name='1800-1899')
        elif 1900 <= year <= 1999:
            books[books.original_publication_year == year].to_excel(xlsxWriter,sheet_name='1900-1999')
        elif 2000 <= year <= 2099:
            books[books.original_publication_year == year].to_excel(xlsxWriter,sheet_name='2000-2099')

xlsxWriter.save()
```
